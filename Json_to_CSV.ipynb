Simple Python script that reads in cards.json and exports a cleaned-up csv database named 'Card_Database.csv' 

```python
import pandas as pd
```

```python
# import cards.json from current directory as pandas dataframe:
card_filename = "cards.json"
with open(card_filename) as card_json_data:
    card_data = pd.read_json(card_json_data)
```

```python
# clean the database:
card_data = card_data.set_index("id")
card_data = card_data.drop(columns=['gifUrl', 'isGeneral', 'localGif', 'slug', 'searchableContent', 'category'])

# reorder columns
card_data = card_data[['name', 'type', 'mana', 'attack', 'hp', 'faction', 'rarity', 'setName', 'description']]

card_data = card_data.fillna('N/A')
```

```python
# export to csv (current directory):
card_data.to_csv('Card_Database.csv')
```
